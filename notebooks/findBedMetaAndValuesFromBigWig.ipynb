```python
import pyBigWig
import numpy
```

```python
bed_file = '../data/genomic_regions/mm9_RefSeq_promoters.bed'
bigWig_file = \
    'http://snpinfo.niehs.nih.gov/ucscview/andy/personal_files/' + \
    'wgEncodeSydhTfbsCh12Ets1IggrabSig.bigWig'
```

```python
bw = pyBigWig.open(bigWig_file)

bin_number = 50
bin_size = 100
window_side = int(bin_number*bin_size/2)

meta_plot = {'bins': [], 'bin_values': []}
bed_totals = {'entry_names': [], 'entry_values': []}

for i in range(bin_number):
    start = -window_side + i * bin_size
    end = start + bin_size - 1
    
    meta_plot['bins'].append((start, end))

with open(bed_file) as f:
    row_values = []
    
    for line in f:
        chrom, start, end, name = line.strip().split()[:4]
        center = int((int(end)-int(start))/2 + int(start))

        bin_values = bw.stats(chrom, center - window_side, center + window_side, nBins = bin_number)
        row_values.append([0 if v is None else v for v in bin_values])
        
        bed_totals['entry_names'].append(name)
        bed_totals['entry_values'].append(sum(row_values[-1]))

meta_plot['bin_values'] = list(numpy.array(row_values).mean(axis=0))
```

```python
print(meta_plot)
```
